## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

```python
# Handle Missing Values
df.isnull().sum()
```

```python
# Fill numeric missing values with mean if dataset has null values
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# Fill categorical missing values with mode if dataset has null values
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled successfully")
```

```python
# Handle Duplicates-Check duplicate rows:
df.duplicated().sum()
```

```python
# Remove duplicates if dataset has duplicates
df = df.drop_duplicates()
```

```python
# Select Numeric columns for Check Outliers
num_cols = df.select_dtypes(include=['int64','float64']).columns

num_cols
```

```python
# Outliers checked with All numerical columns

import matplotlib.pyplot as plt
import seaborn as sns

# Copy dataset
df_before = df.copy()
df_after = df.copy()

# Numeric columns from the dataset
num_cols = [
    'Age',
    'Years of Experience',
    'Salary Before AI',
    'Salary After AI',
    'Work Hours Per Week',
    'Job Satisfaction',
    'Productivity Change %'
]

# IQR Method - Remove Outliers
for col in num_cols:
    Q1 = df_after[col].quantile(0.25)
    Q3 = df_after[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_after = df_after[
        (df_after[col] >= lower) &
        (df_after[col] <= upper)
    ]

# Plot Before & After
plt.figure(figsize=(18,6))

# Before
plt.subplot(1,2,1)
sns.boxplot(data=df_before[num_cols])
plt.title("Before Treating Outliers")
plt.xticks(rotation=45)

# After
plt.subplot(1,2,2)
sns.boxplot(data=df_after[num_cols])
plt.title("After Treating Outliers")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

##### This dataset may not have strong outliers, so before/after looks similar.


```python
# Outliers checked with the column of Salary Before AI & Salary After AI

import matplotlib.pyplot as plt
import seaborn as sns

# Copy dataset
df_before = df.copy()
df_after = df.copy()

# Numeric columns from the dataset
num_cols = [
    'Salary Before AI',
    'Salary After AI'
]

# IQR Method - Remove Outliers
for col in num_cols:
    Q1 = df_after[col].quantile(0.25)
    Q3 = df_after[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_after = df_after[
        (df_after[col] >= lower) &
        (df_after[col] <= upper)
    ]

# Plot Before & After
plt.figure(figsize=(18,6))

# Before
plt.subplot(1,2,1)
sns.boxplot(data=df_before[num_cols])
plt.title("Before Treating Outliers")
plt.xticks(rotation=45)

# After
plt.subplot(1,2,2)
sns.boxplot(data=df_after[num_cols])
plt.title("After Treating Outliers")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

#### **Feature Transformations: Derived Fields**

Created new features to analyze the impact of AI on salaries more effectively. Specifically,calculated the absolute change in salary and the percentage change in salary after AI adoption.

```python
df['Salary Change'] = df['Salary After AI'] - df['Salary Before AI']
df['Salary Change %'] = ((df['Salary After AI'] - df['Salary Before AI']) / df['Salary Before AI']) * 100.
df['Salary Change %'] = df['Salary Change %'].round(2)

print("New derived features added: 'Salary Change' and 'Salary Change %'.")
display(df[['Salary Before AI', 'Salary After AI', 'Salary Change', 'Salary Change %']].head())
```

#### **Univariate Analysis → distribution of single variables (countplot, histogram, boxplot)**

```python
# Countplot

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='AI Adoption Level', data=df,color='cyan',alpha=0.7)
plt.title('Distribution of AI Adoption Level')
plt.xlabel('AI Adoption Level')
plt.ylabel('Count')
plt.show()
```

###### Count plot-Shows the count/frequency of each category in a single categorical variable.
*  This chart shows employee distribution across AI adoption levels.

#### **Bivariate Analysis → relation between two variables (scatterplot, barplot, correlation heatmap)**

```python
# Scatter plot

plt.figure(figsize=(10,6))

plt.scatter(
    df['Salary Before AI'],
    df['Salary After AI'],
    color='orange',
    edgecolors='black',
    alpha=0.6,
    s=60
)

plt.title('Salary Before AI vs Salary After AI')
plt.xlabel('Salary Before AI')
plt.ylabel('Salary After AI')
plt.grid(alpha=0.3)

plt.show()
```

##### **Interpretation of Scatter plot:**

###### Scatterplot-Shows the relationship between two numerical variables.

*   In this chart Shows relationship between Salary Before AI & Salary After AI
*   Helps identify salary trend.
*   Used features: Salary Before AI,Salary After AI






```python
# Bar Plot

avg_productivity = df.groupby('AI Adoption Level')['Productivity Change %'].mean()

plt.figure(figsize=(7,5))
plt.bar(avg_productivity.index, avg_productivity.values,linestyle="--",color='purple',edgecolor='black',alpha=0.8)

plt.title('AI Adoption Level vs Productivity Change %')
plt.xlabel('AI Adoption Level')
plt.ylabel('Average Productivity Change %')

plt.show()
```

##### **Interpretation of Bar plot:**

###### Barplot-Compares values between different categories.

*   In this chart Compares productivity across AI adoption levels.
*   Helps identify whether higher AI adoption improves productivity.
*   Features used: AI Adoption Level, Productivity Change %.







```python
# Correlation Heatmap

plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.show()
```

#### **Interpretation of Correlation Heatmap:**

###### Correlation Heatmap-Shows correlation strength between numerical variables using colors.

*   This heatmap shows correlation between numerical variables.
*   Features- Used All numeric features
*   Dark/high values indicate strong positive relationship & Negative values indicate inverse relationship.
*  From this, we identify which features strongly influence each other.


#### **Multivariate Analysis → relation among 3+ variables (pairplot, grouped analysis, pivot tables, advanced plots)**

```python
# Pair Plot

cols = [
    'Age',
    'Salary Before AI',
    'Productivity Change %',
    'Work Hours Per Week',
    'Salary Change'
]
sns.pairplot(
     df[cols + ['AI Adoption Level']],
    hue='AI Adoption Level'
)
plt.show()
```

##### **Interpretation of Pair Plot:**

###### Pairplot-Shows pairwise relationships between multiple numerical variables at once.

*  Features used:Age, Salary Before AI, Productivity Change %, Work Hours Per Week,Salary Change
*   AI Adoption Level is used as color grouping.
*  It helps identify Which variables move together,Strong/weak relationships,Group patterns by AI adoption,Outliers







```python
# Grouped Analysis

grouped = df.groupby(
    ['Industry', 'AI Adoption Level']
)['Productivity Change %'].mean().unstack().round(2)

print(grouped)
```

##### **Interpretation of Grouped Analysis:**

###### Grouped Analysis-Compares one variable across multiple grouped categories.

*   It helps to identify which industries gain more productivity from AI.
*   Features:Compares Industry + AI Adoption Level + Productivity Change %
*   Shows impact of AI adoption across sectors.

```python
# Pivot Table

pivot = pd.pivot_table(
    df,
    values='Salary After AI',
    index='Education Level',
    columns='Job Status',
    aggfunc='mean'
).round(2)

print(pivot)
```

##### **Interpretation of Pivot Table:**

###### Pivot Table Summarizes data in table format for multi-variable comparison.

*   It Helps to understand salary changes by qualification and employment status.
*   Features- Compares Education Level + Job Status + Salary After AI
*   Shows which education groups benefit more.







```python
# Multi Category Analysis with Pivot table

table = pd.pivot_table(
    df,
    values='Job Satisfaction',
    index='Gender',
    columns='Remote Work',
    aggfunc='mean'
)

print(table)
```

##### **Interpretation of Multi Category Analysis:**

*   It helps to understand satisfaction differences among employee groups.
*   Features:Compares Gender + Remote Work + Job Satisfaction
*   Shows impact of remote work on satisfaction

```python
# Advanced Plot- Bubble Chart:

# Size mapping for Automation Risk
size_map = {
    'Low': 100,
    'Medium': 250,
    'High': 400
}

sizes = df['Automation Risk'].map(size_map)
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    df['Years of Experience'],
    df['Salary After AI'],
    s=sizes,
    c=df['Job Satisfaction'],
    alpha=0.6
)

plt.colorbar(scatter, label='Job Satisfaction')
plt.title('Experience vs Salary After AI (Bubble = Automation Risk)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary After AI')
plt.show()



```

##### **Interpretation of Bubble chart:**

###### Bubble Chart-A scatterplot where bubble size represents an additional variable.

*   It helps understand how experience, salary, automation risk, and satisfaction interact.
*   Features: This advanced bubble chart analyzes 4 variables together-X-axis → Years of Experience,Y-axis → Salary After AI,Bubble Size → Automation Risk,Bubble Color → Job Satisfaction
*   Larger bubbles represent employees with higher automation risk & Color intensity shows differences in job satisfaction.
*   From this, we can identify patterns such as experienced employees earning higher salaries,which groups face higher automation risk,how job satisfaction changes across employee categories.



```python
# Advanced Plot- Violin Plot

plt.figure(figsize=(10,6))
sns.violinplot(
    x='Education Level',
    y='Job Satisfaction',
    hue='Remote Work',
    data=df,
    split=True
)

plt.title('Job Satisfaction by Education Level and Remote Work')
plt.xlabel('Education Level')
plt.ylabel('Job Satisfaction')
plt.show()
```

##### **Interpretation of Violin plot:**

###### Violin Plot-Combines boxplot and density plot to show spread and distribution shape.

*   This violin plot shows the relationship between Education Level, Job Satisfaction, and Remote Work.
*   Features: X-axis → Education Level,Y-axis → Job Satisfaction,Color (Hue) → Remote Work (Yes/No)
*   The width of the violin shows where more employees are concentrated & Wider sections indicate more employees having similar job satisfaction scores.
*   This helps compare satisfaction levels across education groups and remote work categories & It also helps identify spread, distribution shape, and variation among groups.
